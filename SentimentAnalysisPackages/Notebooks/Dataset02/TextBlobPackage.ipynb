Import all necessary packages

```python
import pandas as pd
import numpy as np
from textblob import TextBlob
from sklearn.metrics import accuracy_score
import time
```

Read data from dataset (xlsx file)

```python
data = pd.read_excel (r'/content/drive/MyDrive/SentimentAnalysisPackages /Notebooks/Dataset02/Dataset.xlsx') 
```

Run the model on the dataset

```python
x = data['clean_comment']
y = data['category']
x = np.array(x)
y = np.array(y)

textblob_sentiment_polarity = []

start_time = time.process_time()

for sen in x:

    testimonial = TextBlob(str(sen))
    score = testimonial.sentiment.polarity

    if score > 0:
        textblob_sentiment_polarity.append(1)
    elif score < 0:
        textblob_sentiment_polarity.append(-1)
    else:
        textblob_sentiment_polarity.append(0)

prediction = np.array(textblob_sentiment_polarity)

end_time = time.process_time()
```

Print accuracy score

```python
accuracy_score(y, prediction)
```

Print required time

```python
print("Required Time: ", (end_time-start_time), "seconds")
```
