Import all necessary packages

```python
import pandas as pd
import numpy as np
from textblob import TextBlob
from sklearn.metrics import accuracy_score
import time
```

Read data from dataset (xlsx file)

```python
data = pd.read_excel (r'/content/drive/MyDrive/SentimentAnalysisPackages /Notebooks/Dataset01/dataset.xlsx') 
```

Run the model on the dataset

```python
x = data['Text']
y = data['Label']
x = np.array(x)
y = np.array(y)

textblob_sentiment_polarity = []

start_time = time.process_time()

for sen in x:

    textblob_sentiment_polarity_score = TextBlob(sen).polarity
    if textblob_sentiment_polarity_score > 0:
        textblob_sentiment_polarity.append(1)
    elif textblob_sentiment_polarity_score < 0:
        textblob_sentiment_polarity.append(-1)
    else:
        textblob_sentiment_polarity.append(0)

prediction = np.array(textblob_sentiment_polarity)

end_time = time.process_time()
```

Print accuracy score

```python
accuracy_score(y, prediction)
```

Print required time

```python
print("Required Time: ", (end_time-start_time), "seconds")
```
