Import all necessary packages

```python
import pandas as pd
import numpy as np
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.metrics import accuracy_score
import time
```

Read data from dataset (xlsx file)

```python
data = pd.read_excel (r'/content/drive/MyDrive/SentimentAnalysisPackages /Notebooks/Dataset01/dataset.xlsx') 
```

Run the model on the dataset

```python
x = data['Text']
y = data['Label']
x = np.array(x)
y = np.array(y)

analyzer = SentimentIntensityAnalyzer()

vader_sentiment = []

start_time = time.process_time()

for sen in x:

    score = analyzer.polarity_scores(sen)

    if score['compound'] >= 0.05:
        vader_sentiment.append(1)
    elif score['compound'] <= -0.05:
        vader_sentiment.append(-1)
    else:
        vader_sentiment.append(0)

prediction = np.array(vader_sentiment)

end_time = time.process_time()
```

Print accuracy score

```python
accuracy_score(y, prediction)
```

Print required time

```python
print("Required Time: ", (end_time-start_time), "seconds")
```
